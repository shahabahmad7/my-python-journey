##### Linspace
A useful function for plotting mathematical functions is linspace. Linspace returns evenly spaced numbers over a specified interval.

```python
 # numpy.linspace(start, stop, num = int value)
np.linspace(-5, 5, num = 9)     # num =9 means here that it will give you 9 digits in between -5, -5
```

NumPy uses pseudo-random numbers, which means, the numbers look random but aren't really, they're predetermined.
For consistency, you might want to keep the random numbers you generate similar throughout experiments.
To do this, you can use np.random.seed().

```python
np.random.seed(0)
random = np.random.randint(10, size =(5, 3))
random
```
